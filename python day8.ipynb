```python
#polymorphism same method name different behavior 1.Where same method means speak different behavior barks meowspr
class Animal:
    def speak(self):
        print("Animal sound")
class Dog(Animal):
    def speak(self):
        print("Dog barks")
class Cat(Animal):
    def speak(self):
        print("Cat meows")
animals =[Dog(),Cat()]
for a in animals:
    a.speak()
```

```python
#encapsulation means:hiding data from the class protecting it from the direct acess
#in simple worlda lock the data 
class Bank:
    def __init__(self,money):
        self.__money=money
    def get_money(self):
        return self.__money
b=Bank(1000)
print(b.get_money())
```
