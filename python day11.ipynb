```python
#Nested if 
age=20
citizen=True
if age>=18:
    if citizen:
        print("Eligible to vote")
```

```python
#nested if
age=20
citizen=False
if age>=18:
    if citizen:
        print("eligible to vote")
    else:
        print(" not eligible to vote") 
```

```python
#and nested if
age=20
citizen=True
if age>=18 and citizen:
    print("Eligible")
```
