```python
#Data types in python
"""1.int-Whole number
   2.float-Decimal number
   3.str-Text
   4.bool-True/False
   5.list-collection(changable,muttable)
   6.tuple-collection(not changeable,immutable)
   7.dict-key and value pair"""
chandan=50
salary=10.00000
brother="chandan"
a=5
b=10
print(a<b)
print(a>b)
bike=["ns200","aprilla","royal and field","splendore"]
car=("mercidies","BMW","fourtuner","supra")
mammals={"money":"animal","tiger":"wild animal","cat":"domestic animal"}
print(chandan ,salary ,brother ,bike ,car ,mammals)
```

```python
#Data types in python
"""1.int-Whole number
   2.float-Decimal number
   3.str-Text
   4.bool-True/False
   5.list-collection(changable,muttable)
   6.tuple-collection(not changeable,immutable)
   7.dict-key and value pair"""
chandan=50
salary=10.00000
brother="chandan"
a=5
b=10
print(a<b)
print(a>b)
bike=["ns200","aprilla","royal and field","splendore"]
car=("mercidies","BMW","fourtuner","supra")
mammals={"money":"animal","tiger":"wild animal","cat":"domestic animal"}
print(type(chandan) ,type(salary) ,type(brother) ,type(bike) ,type(car) ,type(mammals))
print(chandan ,salary ,brother ,bike ,car ,mammals)
```
