```python
#Recursion means a function calling itself
#function=call=same function
def count(n):
    if n==0:
        return
    print(n)
    count(n-1)
count(3)
```

```python
class Animal:
    def speak(self):
        print("Animal sounds")
class Dog(Animal):
    def speak(self):
        print("Dog barks")
class cat(Animal):
    def speak(self):
        print("cat meows")
animals=[Dog(),cat()]
for a in animals:
    a.speak()
```

```python
#same method name but different behavior
class Dog:
    def speak(self):
        print("Bark")
class Cat:
    def speak(self):
        print("Hello")
d=Dog()
c=Cat()
d.speak()
c.speak()
```

```python
#constructor __init__()
#A constructor is a special function that runs automatically when an object is created
class Car:
    def __init__(self):
        print("Car created")
c=Car()
```

```python
#encaplulation means wrapping data and methods inside a class
#class=box
#inside a box =data+function
class Car:
    def __init__(self,color):
        self.color=color
    def start(self):
        print("Car started")
c=Car("Red")
print(c.color)
c.start()
```

```python
#one classes can use the properties of another class
#parent class-> child class
class Animal:
    def speak(self):
        print("Animal sounds")
class Dog(Animal):
    pass
d=Dog()
d.speak()
```

```python
class Animal:
    def eat(self):
        print("Eating")
class Dog(Animal):
    def bark(self):
        print("Barking")
d=Dog()
d.eat()
d.bark()
```
