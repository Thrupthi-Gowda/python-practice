```python
class BankAccount:
    def __init__(self,balance):
        self.__balance=balance
    def deposit(self,amount):
        self.__balance+=amount
    def get_balance(self):
        return self.__balance
account=BankAccount(1000)
account.deposit(500)
print(account.get_balance())
```

```python
class Animal:
    def speak(self):
        print("animal makes sound")
class Dog(Animal):
    def bark(self):
        print(" Dog barks")
d=Dog()
d.speak()
d.bark()
```

```python
class Animal:
    def speak(self):
        print("animal  sound")
class Dog(Animal):
    def speak(self):
        print(" Dog barks")
class cat(Animal):
    def speak(self):
        print("cat meows")
animals=[Dog(),cat()]
for a in animals:
    a.speak()
```
