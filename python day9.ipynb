```python
class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
objs=[A(),B()]
for obj in objs:
    obj.show()
    
```

```python
class Animal:
    def speak(self):
        print("Animal Sound")
class Dog(Animal):
    def speak(self):
        print("Dog Bark")
animals=[Animal(),Dog()]
for a in animals:
    a.speak()
```

```python
#abstraction 
from abc import ABC,abstractmethod
class Animal(ABC):
    def sound(self):
        pass
class Dog(Animal):
    def sound(self):
        print("Bark")
d=Dog()
d.sound()
```
